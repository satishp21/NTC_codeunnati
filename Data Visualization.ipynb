```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x,np.cos(x))
ax.axis('equal')

ax.annotate('local maximum',xy=(6.28,1),xytext=(10,4),
           arrowprops=dict(facecolor='black',shrink=0.05))
ax.annotate('local minimum',xy=(5 * np.pi,-1),xytext=(2,-6),
           arrowprops=dict(arrowstyle="->",connectionstyle="angle3,angleA=0,angleB=-90"))
```

# adding annotatae & legend

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10)
y = x^2
z = x^3
 


plt.title("plot drawing")
plt.xlabel("subject")
plt.ylabel("marks")

plt.plot(x,y)
plt.plot(y,z)
plt.plot(x,z)
plt.legend(["race1","race2","race3"],loc=4)


#anotation

plt.annotate(xy=[2,1],text="first Entry", rotation=25)
plt.annotate(xy=[4,6],text="second Entry",rotation=25)
plt.annotate(xy=[8,8],text="third entry",rotation=25)
plt.show()
```

 # Pie chart

```python
#import libraries

import numpy as np
import matplotlib.pyplot as plt

#creating the data set
Branch=["computer","Electrical","civil","mechanical","automobile"]

data = [50,34,40,21,23]
variable = [0.1,0,0.1,0,0]

#create chart

fig = plt.figure(figsize=(10,7))

plt.pie(data, labels=Branch,explode=variable,autopct='%1.1f%%',shadow = True)


plt.show()
```

 # 3D ploting

```python
# import mplot3d toolkit,numpy & matplotlib

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d #3d toolkit lib
```

```python
fig = plt.figure()

#syntax for 3D projection

ax = plt.axes(projection = '3d')

#difining all 3 axes

z = np.linspace(0,1,100)
x = z * np.tan(25*z)
y = z * np.sin(25*z)

#3d ploting

ax.plot3D(x,y,z,'green')
ax.set_title('3d ploting')
plt.show()
```
